# # Pre-Processing

After Analysis you will be clear about what Columns are required and What columns are not required. 

Type, Coverage, AREA, REG , DEV are not required as they don't mean much and would not help us further Hence drop them

Cleaning the Data includes looking at missing Values, If any then Filling them with '0',  deleting them or filling them with the mean values

Converting the type of Columns and Index for easier Access

Renaming the Columns and Re-intializing the Index if needed

Adding an extra feature(Column) called as synthetic feature 

```python
df.drop(columns=['Type','Coverage','AREA', 'REG','DEV'], axis=1, inplace=True) # axis 1 is columns and inplace is to perform drop on the Original
df.head(2)# To Display First two Rows from start
```

```python
df.columns.tolist() # to Make columns values as list so that we can access it and use it later
df.index.tolist()
df.head(2)
```

```python
df.head(2) #After this we can see a DF of two rows, and also can observe that the column names are incomprehensible hence change it
```

```python
df.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region','DevName':'Status'},inplace=True)
df.head(2)
```

```python
df.set_index('Country',inplace=True)# Making one of the Column as an Index
```

```python
df.index.name = None
```

```python
df.head(2)
```

# # Accessing 

This part shows how to access particular columns, particular index in a column , several columns and also how to create a condition

Note: df[''] single bracket means it returns a Series and df[[]] double Bracket means it returns a DataFrame

```python
df.head(2)
```

```python
df['Continent'][:5] # single Coloumn as a Series, Second Bracket [:5] is for scaling which gives first 5 rows , if [:7] first seven rows
```

```python
df[['Continent']][:5] # As a DataFrame
```

```python
df[['Continent', 1982]][:5]  # Selected Columns Display 
```

We can See from Above that to access continent we made it as a string but 1980 as an integer hence it is better to change all the coloumn as type string



```python
df.columns = list(map(str,df.columns))
```

```python
df['1982'][:5]
```

To access Index there are two types 

a. If you know the value name df.loc['value name']

b. if you know the index number df.iloc[28]


```python
df.loc[['India']]
```

```python
df.iloc[[79]]
```

```python
df.loc[['India']][['2010','2011','2012','2013']] # Accessing an Index and Prticular columns for that Index
```

```python
condition = df['Continent'] == 'Asia' # Creating a Condition where only Asian continents Values are Displayed
```

```python
df[condition]
```

```python
condition2 = df['Status'] == 'Developed regions'
```

```python
df[condition][condition2]#Two Conditions In Asia which is the developed country
```

```python
#Another way of Accessing 

df[(df['Continent'] == 'Asia') & (df['Status'] == 'Developed regions')] # USe & || for (and or) in Pandas
```
